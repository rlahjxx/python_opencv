```python
img_lists = os.listdir('../opencv_day 1/fig/my_images/')

img_files = []
for i in img_lists:
    img_file = '../opencv_day 1/fig/my_images/' + i
    img_files.append(img_file)

cv2.namedWindow('scene', cv2.WINDOW_NORMAL)
cv2.setWindowProperty('scene', cv2.WND_PROP_FULLSCREEN, cv2.WINDOW_FULLSCREEN)

idx = 0
while True:
    img = cv2.imread(img_files[idx])
    
    if img is None:
        print('image read failed')
        break
        
    cv2.imshow('scene', img)
    
    key = cv2.waitKey(3000)
    if key == 27 or key == ord('q'):
        break
        
    idx += 1
    
    if idx >= len(img_files):
        idx = 0
cv2.destroyAllWindows()
```

```python
img1 = cv2.imread('./fig/cow.png')
img2 = cv2.imread('./fig/green.png')

img1_grey = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
retval, img1_mask = cv2.threshold(img1_grey, 244, 255, cv2.THRESH_BINARY_INV)

h, w = img1.shape[:2]
img2_resize = cv2.resize(img2, (w,h), interpolation = cv2.INTER_LINEAR)

cv2.copyTo(img1, img1_mask, img2_resize)

cv2.imshow('img1', img1)
cv2.imshow('img2', img2)
cv2.imshow('img1_grey', img1_grey)
cv2.imshow('img1_mask', img1_mask)
cv2.imshow('img2_resize', img2_resize)


cv2.waitKey()
cv2.destroyAllWindows()
```

```python
src = cv2.imread('./fig/puppy.bmp')
alpha_image = cv2.imread('./fig/imgbin_sunglasses_1.png', cv2.IMREAD_UNCHANGED)

alpha_image = cv2.resize(alpha_image, (300,150), interpolation=cv2.INTER_AREA)

h, w = alpha_image.shape[:2]
crop = src[120:120+h, 220:220+w]

mask = alpha_image[:,:,-1]
glass = alpha_image[:,:,0:3]

cv2.copyTo(glass, mask, crop)
# crop[mask > 0] = (0,0,255)

cv2.imshow('src', src)
cv2.imshow('mask', mask)
cv2.imshow('glass', glass)
cv2.imshow('crop', crop)

cv2.waitKey()
cv2.destroyAllWindows()
```

## 직선, 원 그려보기

```python
img = np.ones((600,1200,3), np.uint8)*255

# cv2.line(img, pt1, pt2, color, thickness = None, lineType = None, shift = None) -> img
# flags
    # img:그림을 그릴 영상
    # pt1, pt2: 직선의 시작점, 끝점
    # color: 직선의 칼라 (B,G,R)의 튜플
    # thinkness: 선두께, 기본은= 1
    # lineType: cv2.LINE_4, cv2.LINE_8, cv2.LINE_AA
    
cv2.line(img, (100,30), (350,30), (0,0,255), 8, cv2.LINE_4)
cv2.line(img, (350,30), (250,300), (0,0,255), 8, cv2.LINE_AA)
cv2.arrowedLine(img, (450,30), (450,300), (255,0,0), 8, cv2.LINE_4)

# cv2.rectangle(img, pt1, pt2, color, thickness = None, lineType = None) -> img
   # pt1 :좌측 상단,  pt2: 우측하단
# cv2.rectangle(img, rect, color, thickness = None, lineType = None) -> img
    # rect: 사각형의 위치 정보 (x, y, w, h)

cv2.rectangle(img, (100,400), (450,500), (0,0,255), 8, cv2.LINE_4)
# cv2.rectangle(img, (120,420), (430,480), (0,255,0), -1, cv2.LINE_4)
cv2.rectangle(img, (120, 420, 310, 60), (0,255,0), -1)

# cv2.circle(img, center, radius, color, thickness = None, lineType = None) -> img
    # center: 원의 중심좌표 (x, y)
    # radius : 원의 반지름

cv2.circle(img, (800, 200), 100, (0,0,255), 5, cv2.LINE_AA)

# cv2.ellips(img, center, axes, angle, startAngle, endAngle, color[, thickness[, lineType[, shift]]]) -> img
    # center: 원의 중심좌표 (x, y)
    # axis: 축의 반지름(x, y)
    # angle: 타원의 기울기 (예, 10, 오른쪽으로 10도 기울어짐)
    # startAngle: 타원을 그리는 시작 각도 (3시 방향이 0도)
    # endAngle: 타원을 그리는 종료 각도
    
cv2.ellipse(img, (800,200), (50,200), 10, 0, 270, (255,0,255), 5, cv2.LINE_AA)

# putText(img, text, org, fontFace, fontScale, color[, thickness[, lineType[, bottomLeftOrigin]]])
cv2.putText(img, 'Hello opencv!', (900,400), cv2.FONT_HERSHEY_COMPLEX, 
           1, (123, 222, 44), 2, cv2.LINE_AA)

cv2.imshow('image', img)

cv2.waitKey()
cv2.destroyAllWindows()
```
