```python
import numpy as np
import cv2
import sys

print(cv2.__version__)
```

## 이미지 열기 기본 형태

```python
# cv2.IMREAD_COLOR : BGR color로 읽기
# cv2.IMREAD_GRAYSCALE : 그레이 color로 읽기
# cv2.IMREAD_UNCHANGED : 파일 속성대로 읽기
# retval : numpy.ndarray로 반환

img = cv2.imread('./fig/puppy.bmp', cv2.IMREAD_COLOR)

if img is None:
    print('Image read failed')
    sys.exit()

cv2.namedWindow('puppy')
cv2.imshow('puppy', img)
cv2.waitKey() # 숫자를 넣어주면 그만큼 있다가 꺼짐

cv2.destroyAllWindows()
```

```python
# cv2.IMREAD_COLOR : BGR color로 읽기
# cv2.IMREAD_GRAYSCALE : 그레이 color로 읽기
# cv2.IMREAD_UNCHANGED : 파일 속성대로 읽기
# retval : numpy.ndarray로 반환

img = cv2.imread('./fig/puppy.bmp', cv2.IMREAD_GRAYSCALE)

# resize(src, dsize[, dst[, fx[, fy[, interpolation]]]]) -> dst
img_resize = cv2.resize(img, (480, 640), interpolation = cv2.INTER_LINEAR)

# cv2.imwrite(filename, img, params = None) => retval

# filename : 저장할 파일 이름, img : 저장할 자료(numpy.ndarray)
# params : 파일저장 옵션, retval : 성공 True, 실패 False

cv2.imwrite('./fig/resize.png', img_resize)

if img is None:
    print('Image read failed')
    sys.exit()
    
# cv2.namedWindow(winname, flags=None) -> None
# flags:
# cv2.WINDOW_NORMAL : 영상 크기를 창 크기에 맞게 변경
# cv2.WINDOW_AUTOSIZE : 창 크기를 영상 크기에 맞게 변경

cv2.namedWindow('puppy', cv2.WINDOW_AUTOSIZE)
cv2.imshow('puppy', img)

# cv2.imshow('resize', img_resize)
cv2.waitKey()

# while True:
#     key = cv2.waitKey()
    
#     if key == 27 or key == ord('q'):
#         break

cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import sys
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('./fig/puppy.bmp', cv2.IMREAD_GRAYSCALE)
imgBGR = cv2.imread('./fig/puppy.bmp', cv2.IMREAD_COLOR)
# cvtColor(src, code[, dst[, dstCn]])
imgRGB = cv2.cvtColor(imgBGR, cv2.COLOR_BGR2RGB)

if img is None:
    print('image read failed')
    sys.exit()

plt.figure(figsize=(10,10))
plt.subplot(131), plt.imshow(imgRGB), plt.axis('off')
plt.subplot(132), plt.imshow(imgBGR), plt.axis('off')
plt.subplot(133), plt.imshow(img, cmap='gray'), plt.axis('off')
plt.show()  # => BGR과 RGB 의 구분 알아야 함
    
# cv2.namedWindow('puppy', cv2.WINDOW_AUTOSIZE)
# cv2.imshow('imgBGR', imgBGR)
# cv2.imshow('imgRGB', imgRGB)
# cv2.imshow('img', img)

# while True:
#     key = cv2.waitKey()
#     if key == 27 or key == ord('q'):
#         break
        
# cv2.destroyAllWindows()
```
